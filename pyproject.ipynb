```python
#avg birtyear
import pandas as pd

df = pd.read_csv("student133.csv")

# Convert birthdate column to datetime
df['birthdate'] = pd.to_datetime(df['birthdate'])

# Extract just the year from the birthdate
df['birth_year'] = df['birthdate'].dt.year

# Calculate the average (mean) birth year
avg_birth_year = df['birth_year'].mean()
print(f"The average birth year is: {avg_birth_year:.2f}")
```

```python
#all female born after 2000
import pandas as pd

df = pd.read_csv("student133.csv")

df['birthdate'] = pd.to_datetime(df['birthdate'])

# Filter: gender == 'Female' AND birthdate > '2002-12-31'
filtered_students = df[(df['gender'] == 'Female') & (df['birthdate'] > '2002-12-31')]

print("Female students born after 2002:")
print(filtered_students)
```

```python
# Identify students who live in the same city they were from (district vs. city mismatch/match)

import pandas as pd

df = pd.read_csv("student133.csv")

# We'll assume district is "from" and city is "current city" based on your CSV field meaning.
same_location = df[df['district'] == df['city']]
moved_location = df[df['district'] != df['city']]

print("Students living in the same district and city:")
print(same_location)

print("\nStudents who have moved to a different city:")
print(moved_location)
```

```python
#youngest and oldest
import pandas as pd

df = pd.read_csv("student133.csv")

df['birthdate'] = pd.to_datetime(df['birthdate'])

# Sort by birthdate ascending (oldest at the top if you consider earliest birthdate)
sorted_df = df.sort_values(by='birthdate')

oldest_student = sorted_df.iloc[0]
youngest_student = sorted_df.iloc[-1]

print("Oldest student:\n", oldest_student)
print("\nYoungest student:\n", youngest_student)
```

```python
# Add a new column showing the studentâ€™s age

import pandas as pd
from datetime import datetime

df = pd.read_csv("student133.csv")

df['birthdate'] = pd.to_datetime(df['birthdate'])

# Calculate approximate age by subtracting birth_year from current year
current_year = datetime.now().year
df['age'] = current_year - df['birthdate'].dt.year

print("DataFrame with new 'age' column:")
print(df[['fullname', 'birthdate', 'age']])
```

```python
#group student by birth year
import pandas as pd

df = pd.read_csv("student133.csv")
df['birthdate'] = pd.to_datetime(df['birthdate'])

# Extract the year
df['birth_year'] = df['birthdate'].dt.year

# Group by the birth_year and count
students_per_year = df.groupby('birth_year').size().reset_index(name='count')
print("Number of students in each birth year:")
print(students_per_year)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the CSV file
df = pd.read_csv("student133.csv")

# --- Gender Distribution ---
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(6,4))
gender_counts.plot(kind='bar', color=['skyblue','lightcoral'])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.xticks(rotation=0)
plt.show()
```

```python
# --- District Distribution ---
import pandas as pd
import matplotlib.pyplot as plt

district_counts = df['district'].value_counts()
plt.figure(figsize=(10,5))
district_counts.plot(kind='bar', color='mediumseagreen')
plt.title("District Distribution")
plt.xlabel("District")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert birthdate to datetime and extract year
import pandas as pd
import matplotlib.pyplot as plt

df['birthdate'] = pd.to_datetime(df['birthdate'])
df['birth_year'] = df['birthdate'].dt.year

plt.figure(figsize=(8,5))
plt.hist(df['birth_year'], bins=range(df['birth_year'].min(), df['birth_year'].max()+1),
         color='slateblue', edgecolor='black')
plt.title("Birth Year Distribution")
plt.xlabel("Year of Birth")
plt.ylabel("Number of Students")
plt.show()
```
